```python
def scale(x):
    """
    Scales the input data by subtracting the mean and dividing by the standard deviation 
    
    Inputs: x: a n-dimensional tensor
    
    Returns: a n-dimensional tensor of same shape as x
    """

    return (x - np.mean(x)) / np.std(x)
```

```python
def preprocess(x):
    """
    Cuts the input 75% along the z-axis to remove laminar layers
    
    Inputs: x: a 3-dimensional tensor
    
    Returns: a 3-dimensional tensor of shape (length, width, 75% * height)
    """
    
    return x[:,:,:int(0.75*x.shape[2])]
```

```python
# Read data
os.chdir(path_to_folder + "/Re1243Fr20/coarse8")
u = preprocess(scale(loadmat('u_F_xyz_T1.mat')["u_F"]))
v = preprocess(scale(loadmat('v_F_xyz_T1.mat')["v_F"]))
w = preprocess(scale(loadmat('w_F_xyz_T1.mat')["w_F"]))
tau_12 = preprocess(scale(loadmat('tau12_F_xyz_T1.mat')["tau12"]))
tau_13 = preprocess(scale(loadmat('tau13_F_xyz_T1.mat')["tau13"]))
tau_23 = preprocess(scale(loadmat('tau23_F_xyz_T1.mat')["tau23"]))
tke = preprocess(scale(loadmat("TKE_F_xyz_T1.mat")["TKE_F"]))
theta = preprocess(scale(loadmat("theta_F_xyz_T1.mat")["theta_F"]))
p = preprocess(scale(loadmat("pstar_F_xyz_T1.mat")["pstar_F"]))
```

### 1. Generae Full Input Dataset

```python
x = np.array([u, v, w, tke, theta, p])
x = np.transpose(x, [1, 2, 3, 0])
x = np.pad(x, ((3,3), (3,3), (3,3), (0,0)), 'constant', constant_values = 0)
```

```python
# Generate randomized dataset shards
sample = []
size = 3
y_tau_12 = []
y_tau_13 = []
y_tau_23 = []
for i in tqdm_notebook(range(size, x.shape[0] - size)):
    for j in range(size, x.shape[1] - size):
        for k in range(size, x.shape[2] - size):
            sample.append(x[i - size: i + size + 1, j - size: j + size + 1, k - size: k + size + 1, :])
            y_tau_12.append(tau_12[i - size][j - size][k - size])
            y_tau_13.append(tau_13[i - size][j - size][k - size])
            y_tau_23.append(tau_23[i - size][j - size][k - size])
```

```python
data = np.stack(sample, axis = 0)
print(data.shape)

y_tau_12 = np.array(y_tau_12)
y_tau_13 = np.array(y_tau_13)
y_tau_23 = np.array(y_tau_23)

del sample
del u
del v
del w
del p
del theta
del tke
del tau_12
del tau_13
del tau_23
gc.collect()
```

### 2. Build Model 

```python
class DNN(object):
    
    def __init__(self, activation, x_train, y_train, epochs, batch_size, input_shape, val_split, initializer):
        self.activation = activation
        self.x_train = x_train
        self.y_train = y_train
        self.epochs = epochs
        self.batch_size = batch_size
        self.input_shape = input_shape
        self.val_split = val_split
        self.initializer = initializer
        
        pass 
    
    def create_model(self):
        """
        Creates sequential keras model

        Inputs: self

        Returns: a sequential keras model with the parameters defined in __init__
        """
        
        model = keras.Sequential([
            tf.keras.layers.Flatten(input_shape = self.input_shape),
            tf.keras.layers.Dense(128, 
                             activation = self.activation,
                             kernel_initializer = self.initializer),
            tf.keras.layers.Dropout(0.2),
            tf.keras.layers.Dense(64, 
                             activation = self.activation,
                             kernel_initializer = self.initializer),
            tf.keras.layers.Dropout(0.2),
            tf.keras.layers.Dense(32, 
                             activation = self.activation,
                             kernel_initializer = self.initializer),
            tf.keras.layers.Dense(1)
        ])
        
        return model
    
    def run_model(self):
        """
        Creates, compiles and runs the model

        Inputs: self

        Returns: a sequential keras model with the parameters defined in __init__ and its history
        """
        
        model = self.create_model()
        model.compile(optimizer = tf.train.AdamOptimizer(), 
              loss = "mse")
            
        history = model.fit(self.x_train, self.y_train, 
                    epochs = self.epochs, 
                    validation_split = self.val_split, 
                    batch_size =  self.batch_size,
                    verbose = 0)
        
        return history, model
```

### 3. Run Model

```python
out = [["Input Dataset", "Tau_12", "Tau_13", "Tau_23"]]
inputs = ["(u,v,w)", "(u,v,w,tke)", "(u,v,w,theta)", "(u,v,w,p)"]
taus = [(y_tau_12, "tau_12"), (y_tau_13, "tau_13"), (y_tau_23, "tau_23")]
i = 1

for dataset in inputs:
    if dataset == "(u,v,w)":
        x = np.delete(data, (3,4,5), 4)
    elif dataset == "(u,v,w,tke)":
        x = np.delete(data, (4,5), 4)
    elif dataset == "(u,v,w,theta)":
        x = np.delete(data, (3,5), 4)
    else:
        x = np.delete(data, (3,4), 4)
    
    # Define model parameters
    activation = tf.nn.relu
    epochs = 50
    batch_size = 1000
    initializer = None
    input_shape = (x.shape[1], x.shape[2], x.shape[3], x.shape[4])
    val_split = 0.2

    fig = plt.figure(figsize = (20, 24))
    correlation = []
    for (y, name) in taus:
        #Create train and test data
        mask =  np.random.rand(x.shape[0]) < 0.80
        x_train, x_test = x[mask,:,:,:,:], x[~mask,:,:,:,:]
        y_train, y_test = y[mask], y[~mask]
        
        # Create network instance and fit model
        model = DNN(activation, x_train, y_train, epochs, batch_size, input_shape, val_split, initializer)
        history, model = model.run_model()

        del x_train
        del y_train
        gc.collect()
        
        # Plot R2 and corr evolution
        fig.add_subplot(4, len(taus), i)
        plt.plot(history.epoch, np.array(history.history['loss']), label = 'Train Loss')
        plt.plot(history.epoch, np.array(history.history['val_loss']), label = 'Val loss')
        plt.xlabel('Epoch')
        plt.ylabel('Mean Square Error')
        plt.title(dataset)
        plt.legend()
        
        # Get test corr and R2
        y_pred = model.predict(x_test).flatten()
        y_true = y_test
        correlation.append(np.corrcoef(y_pred, y_true)[0, 1])

        del x_test
        del y_true
        del y_pred
        del y_test
        del model
        del y
        gc.collect()
        i += 1

    out.append([dataset, str(np.round(correlation[0], 4)), str(np.round(correlation[1], 4)), str(np.round(correlation[2], 4))])

plt.show()
display(HTML(tabulate.tabulate(out, tablefmt = 'html')))
```
